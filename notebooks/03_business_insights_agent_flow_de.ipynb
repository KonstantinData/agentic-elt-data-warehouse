## ğŸ“Š Pipeline-AusfÃ¼hrungsanalyse

**Was passiert hier:**
- Analyse der vollstÃ¤ndigen Pipeline-AusfÃ¼hrungszusammenfassung
- ÃœberprÃ¼fung von Performance-Metriken und AusfÃ¼hrungszeiten
- Verstehen des Datenflusses durch Bronze â†’ Silver â†’ Gold Layer

**Warum wichtig:**
Der Business Insights Agent nutzt Pipeline-AusfÃ¼hrungsdaten, um Kontext Ã¼ber DatenaktualitÃ¤t, Verarbeitungserfolg und Systemperformance in seinen Reports zu liefern.

```python
# Pipeline-AusfÃ¼hrungszusammenfassung analysieren
summary_json_path = reports_dir / "summary_report.json"
if summary_json_path.exists():
    with open(summary_json_path, 'r') as f:
        summary_report = json.load(f)
    
    print("ğŸ“Š PIPELINE-AUSFÃœHRUNGSANALYSE")
    print("=" * 50)
    
    print(f"\nâ±ï¸ AusfÃ¼hrungs-Timeline:")
    print(f"  Run ID: {summary_report.get('run_id')}")
    print(f"  Gestartet: {summary_report.get('started_utc')}")
    print(f"  Beendet: {summary_report.get('ended_utc')}")
    print(f"  Dauer: {summary_report.get('duration_s', 0):.2f} Sekunden")
    
    # Schritt-fÃ¼r-Schritt Analyse
    steps = summary_report.get('step_results', [])
    print(f"\nğŸ”„ Pipeline-Schritte ({len(steps)} gesamt):")
    
    successful_steps = 0
    failed_steps = 0
    total_duration = 0
    
    for step in steps:
        status = step.get('status', 'unknown')
        duration = step.get('duration_s', 0)
        total_duration += duration
        
        status_icon = "âœ…" if status == "success" else "â­ï¸" if status == "skipped" else "âŒ"
        print(f"  {status_icon} {step.get('name')}: {status} ({duration:.2f}s)")
        
        if status == "success":
            successful_steps += 1
        elif status == "failed":
            failed_steps += 1
    
    # Layer Run IDs
    print(f"\nğŸ—ï¸ Daten-Layer Run IDs:")
    print(f"  Bronze: {summary_report.get('bronze_run_id')}")
    print(f"  Silver: {summary_report.get('silver_run_id')}")
    print(f"  Gold: {summary_report.get('gold_run_id')}")
    
    # Performance-Zusammenfassung
    print(f"\nğŸ“ˆ Performance-Zusammenfassung:")
    print(f"  Erfolgreiche Schritte: {successful_steps}/{len(steps)}")
    print(f"  Fehlgeschlagene Schritte: {failed_steps}")
    print(f"  Keine neuen Daten erkannt: {summary_report.get('no_new_data', False)}")
else:
    print("âŒ Pipeline-Zusammenfassungsreport nicht gefunden")
```

## ğŸ‘¥ Stakeholder-spezifische Reports

**Was passiert hier:**
- Analyse team-spezifischer Reports fÃ¼r verschiedene Business-Funktionen
- ÃœberprÃ¼fung, wie der Agent Insights fÃ¼r Marketing-, Sales-, Product-, IT- und Customer Service-Teams anpasst
- Verstehen rollenbasierter Business Intelligence Bereitstellung

**Zielgerichtete Intelligence:**
Der Business Insights Agent erstellt maÃŸgeschneiderte Reports, die sich auf Metriken und Insights konzentrieren, die fÃ¼r jede Business-Funktion relevant sind.

```python
# Stakeholder-spezifische Reports analysieren
team_reports_dir = reports_dir / "team_reports"
if team_reports_dir.exists():
    team_report_files = list(team_reports_dir.glob("*.md"))
    
    print(f"ğŸ‘¥ STAKEHOLDER-SPEZIFISCHE REPORTS ANALYSE")
    print("=" * 60)
    
    print(f"\nğŸ“‹ VerfÃ¼gbare Team-Reports ({len(team_report_files)}):")
    
    team_analysis = {}
    
    for report_file in team_report_files:
        team_name = report_file.stem.replace('_team_report', '').replace('_', ' ').title()
        
        with open(report_file, 'r', encoding='utf-8') as f:
            content = f.read()
        
        # Report-Inhalt analysieren
        lines = content.split('\n')
        word_count = len(content.split())
        sections = content.count('#')
        
        team_analysis[team_name] = {
            'lines': len(lines),
            'words': word_count,
            'sections': sections,
            'content': content
        }
        
        print(f"\nğŸ“Š {team_name} Team Report:")
        print(f"  LÃ¤nge: {word_count:,} WÃ¶rter, {len(lines)} Zeilen")
        print(f"  Abschnitte: {sections}")
        
        # Report-Vorschau zeigen
        preview_lines = []
        for line in lines[:12]:  # Erste 12 Zeilen
            if line.strip():
                preview_lines.append(line)
            if len(preview_lines) >= 6:  # Max 6 relevante Zeilen
                break
        
        print(f"  Vorschau:")
        for line in preview_lines:
            print(f"    {line}")
        print(f"    ... (gekÃ¼rzt)")
    
    # Zusammenfassung der Team-Reports
    if team_analysis:
        total_words = sum(t['words'] for t in team_analysis.values())
        avg_sections = sum(t['sections'] for t in team_analysis.values()) / len(team_analysis)
        
        print(f"\nğŸ“ˆ Team-Reports Zusammenfassung:")
        print(f"  Gesamt Reports: {len(team_analysis)}")
        print(f"  Gesamt Inhalt: {total_words:,} WÃ¶rter")
        print(f"  Durchschnittliche Abschnitte pro Report: {avg_sections:.1f}")
        print(f"  Abgedeckte Teams: {', '.join(team_analysis.keys())}")
else:
    print("âŒ Team-Reports nicht gefunden")
```

## ğŸ” DatenqualitÃ¤t und Pipeline-Gesundheit

**Was passiert hier:**
- Analyse von DatenqualitÃ¤tsmetriken aus Gold Layer Data Marts
- ÃœberprÃ¼fung von Pipeline-Performance und Gesundheitsindikatoren
- Bewertung der Business Intelligence DatenzuverlÃ¤ssigkeit

**QualitÃ¤tssicherung:**
Das Verstehen der DatenqualitÃ¤t ist entscheidend fÃ¼r das Vertrauen in Business-Insights und fundierte Entscheidungen basierend auf den Analytics.

```python
# DatenqualitÃ¤t fÃ¼r Business-Insights analysieren
gold_data_dir = gold_dir / "data"
if gold_data_dir.exists():
    gold_files = list(gold_data_dir.glob("*.csv"))
    
    print(f"ğŸ” DATENQUALITÃ„T UND PIPELINE-GESUNDHEIT")
    print("=" * 50)
    
    if gold_files:
        total_records = 0
        total_columns = 0
        data_quality_summary = {}
        
        print(f"\nğŸ“Š Gold Layer DatenqualitÃ¤tsbewertung:")
        
        for gold_file in gold_files:
            df = pd.read_csv(gold_file)
            records = len(df)
            columns = len(df.columns)
            total_records += records
            total_columns += columns
            
            # DatenqualitÃ¤tsmetriken berechnen
            null_count = df.isnull().sum().sum()
            total_cells = records * columns
            null_percentage = (null_count / total_cells * 100) if total_cells > 0 else 0
            completeness = 100 - null_percentage
            
            # Numerische Spalten fÃ¼r Business-Metriken identifizieren
            numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
            
            data_quality_summary[gold_file.name] = {
                'records': records,
                'columns': columns,
                'null_percentage': null_percentage,
                'completeness': completeness,
                'numeric_columns': len(numeric_cols),
                'business_metrics_available': len(numeric_cols) > 0
            }
            
            quality_icon = "âœ…" if completeness >= 95 else "âš ï¸" if completeness >= 80 else "âŒ"
            print(f"  {quality_icon} {gold_file.name}:")
            print(f"    DatensÃ¤tze: {records:,}")
            print(f"    VollstÃ¤ndigkeit: {completeness:.1f}%")
            print(f"    Business-Metriken: {len(numeric_cols)} numerische Spalten")
        
        # GesamtqualitÃ¤tsbewertung
        avg_completeness = sum(q['completeness'] for q in data_quality_summary.values()) / len(data_quality_summary)
        tables_with_metrics = sum(1 for q in data_quality_summary.values() if q['business_metrics_available'])
        
        print(f"\nğŸ“ˆ Gesamte DatenqualitÃ¤ts-Zusammenfassung:")
        print(f"  Gesamt Business-DatensÃ¤tze: {total_records:,}")
        print(f"  Gesamt Data Marts: {len(gold_files)}")
        print(f"  Durchschnittliche VollstÃ¤ndigkeit: {avg_completeness:.1f}%")
        print(f"  Tabellen mit Business-Metriken: {tables_with_metrics}/{len(gold_files)}")
        
        # QualitÃ¤tsbewertung
        if avg_completeness >= 95:
            quality_rating = "Ausgezeichnet âœ…"
        elif avg_completeness >= 85:
            quality_rating = "Gut âš ï¸"
        elif avg_completeness >= 70:
            quality_rating = "Befriedigend âš ï¸"
        else:
            quality_rating = "Mangelhaft âŒ"
        
        print(f"\nğŸ¯ Business Intelligence DatenqualitÃ¤ts-Rating: {quality_rating}")
        print(f"  Empfehlung: {'Daten sind zuverlÃ¤ssig fÃ¼r Business-Entscheidungen' if avg_completeness >= 85 else 'DatenqualitÃ¤tsprobleme vor kritischen Entscheidungen Ã¼berprÃ¼fen'}")
    else:
        print(f"\nâš ï¸ Keine Gold Layer Daten fÃ¼r QualitÃ¤tsbewertung gefunden")
else:
    print("âŒ Gold Daten nicht verfÃ¼gbar fÃ¼r QualitÃ¤tsanalyse")
```

```python
# Business Insights Agent AusfÃ¼hrungslog analysieren
business_log_path = orchestrator_dir / "logs" / "business_insights.log"
if business_log_path.exists():
    with open(business_log_path, 'r', encoding='utf-8') as f:
        log_content = f.read()
    
    print("ğŸ“‹ BUSINESS INSIGHTS AGENT AUSFÃœHRUNGSLOG")
    print("=" * 60)
    
    # Zeige aktuelle Log-EintrÃ¤ge
    lines = log_content.split('\n')
    print(f"\nğŸ“„ Aktuelle Log-EintrÃ¤ge (letzte 25 Zeilen):")
    for line in lines[-25:] if len(lines) > 25 else lines:
        if line.strip():
            print(f"  {line}")
    
    # Log-Analyse
    error_lines = [line for line in lines if 'ERROR' in line.upper()]
    warning_lines = [line for line in lines if 'WARNING' in line.upper()]
    info_lines = [line for line in lines if 'INFO' in line.upper()]
    
    print(f"\nğŸ“Š Log-Analyse Zusammenfassung:")
    print(f"  Gesamt Log-Zeilen: {len(lines)}")
    print(f"  Info-Nachrichten: {len(info_lines)}")
    print(f"  Warnungen: {len(warning_lines)}")
    print(f"  Fehler: {len(error_lines)}")
    
    # Zeige kritische Probleme falls vorhanden
    if error_lines:
        print(f"\nâŒ Kritische Probleme gefunden:")
        for error in error_lines[-3:]:  # Letzte 3 Fehler
            print(f"  {error.strip()}")
    
    if warning_lines:
        print(f"\nâš ï¸ Warnungen gefunden:")
        for warning in warning_lines[-3:]:  # Letzte 3 Warnungen
            print(f"  {warning.strip()}")
    
    # Agent-Performance Indikatoren
    llm_calls = log_content.lower().count('llm') + log_content.lower().count('openai')
    report_generations = log_content.lower().count('report') + log_content.lower().count('generated')
    
    print(f"\nğŸ¤– Agent-Performance Indikatoren:")
    print(f"  LLM-Interaktionen: ~{llm_calls} (geschÃ¤tzt)")
    print(f"  Report-Generierungen: ~{report_generations} (geschÃ¤tzt)")
    print(f"  AusfÃ¼hrungsstatus: {'âœ… Erfolgreich' if len(error_lines) == 0 else 'âŒ Probleme erkannt'}")
else:
    print("âŒ Business Insights Agent Log nicht gefunden")
```
